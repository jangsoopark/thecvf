```python
from collections import Counter

import pandas as pd

import re
import os


project_root = os.path.dirname(os.getcwd())
```

```python
df = pd.read_json(os.path.join(project_root, 'assets/wacv-2023-main.json'))
```

```python
df.loc[:, ('title', 'abstract', 'author')]
```

```python
titles = df.loc[:, 'title']

count_of_keyword_in_title = Counter()
keyword_in_title = set()
for title in titles:
    title = re.sub('[^0-9a-zA-Z- ]', ' ', title)
    count_of_keyword_in_title.update(title.split())
    keyword_in_title.update(title.split())
```

```python
sorted(count_of_keyword_in_title.items(), key=lambda x: x[1], reverse=True)
```
